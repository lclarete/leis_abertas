```python
# READ FILE 2: read file from "Leis Abertas"

url_cidade = pd.read_csv('Leis Abertas - Cidades.csv')

new_columns = ['municipio', 'uf', 
               'CodMun_old', 
               'lexml', 
               'site_prefeitura',
               'url_legislativo_2', 
               'sistema_tipo', 
               'sistema_proposicao',
               'sistema_legislacao', 
               'Apoio Interlegis']

url_cidade.columns = new_columns
```

```python
url_cidade.head()
```

# feature extraction

```python
# DF 1: url_cidade
uf_code_dict = {'RO': 11, 
                'AC': 12, 
                'AM': 13, 
                'RR': 14, 
                'PA': 15, 
                'AP': 16, 
                'TO': 17, 
                'MA': 21, 
                'PI': 22, 
                'CE': 23, 
                'RN': 24, 
                'PB': 25, 
                'PE': 26, 
                'AL': 27, 
                'SE': 28, 
                'BA': 29, 
                'MG': 31, 
                'ES': 32, 
                'RJ': 33, 
                'SP': 35, 
                'PR': 41, 
                'SC': 42, 
                'RS': 43, 
                'MS': 50, 
                'MT': 51, 
                'GO': 52, 
                'DF': 53}
```

```python
# swift names to codes
url_cidade['uf_code'] = url_cidade.uf.map(uf_code_dict)

# normilize the CodNun column
url_cidade['CodMun'] = url_cidade.uf_code.multiply(100000) + url_cidade.CodMun_old
```

```python
# merge the data sets using CodNum -- IBGE code as a matcher col
df_total = url_interlegis.merge(url_cidade, on=['CodMun'], how='inner')
```

```python
# delete unuseful columns
df_url = df_total.drop(['match_var', 
                        'uf_y', 'uf_code_y', 
                        'CodMun_old', 'uf_code_multi', 
                        'uf_x', 
                       'municipio'], 
                       axis=1)
```

```python
print(df_url.shape)
print(df_url.columns)
df_url.head()
```

# check if the urls are up

```python
import requests

def url_ok(col_url):
    r = requests.head(col_url)
    check = r.status_code == 200
    return(check)
```

```python
import requests

def url_check(col_url):
    try:
        r = requests.get(col_url, timeout=10)
        return (r.status_code == 200)
    except requests.exceptions.RequestException as e:
        return('Read timed out')
```

```python
df_url['url_check'] = df_url.url_legislativo.apply(url_check)
```
